```python
# Author: Arash 
# Date Published: May 19, 2023
# Purpose: Plotting the PDF of the chi-square distribution for different values of K (A1 Q1(c))
# Language: Python

# Importing required packages 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

K_values = [1, 2, 3, 10, 100] # Defined in problem statment
colors = ['blue', 'red', 'green', 'orange', 'purple'] # Custom color coding for plot

x = np.linspace(0, 140, 1000)  # Range of values for y

plt.figure(figsize=(10, 6))  # Adjust the figure size

for K, color in zip(K_values, colors):
    y = chi2.pdf(x, K)  # PDF of y using chi2 function of scipy package 

    plt.plot(x, y, color=color, label=f'K = {K}')

# Plot settings
plt.xlabel('y')
plt.ylabel('PDF')
plt.title('PDF of y for different values of K')
plt.xlim(0, 140)  # Adjusted x-axis limits
plt.ylim(0, 0.35)  # Adjusted y-axis limits
plt.legend()
plt.show()
```
